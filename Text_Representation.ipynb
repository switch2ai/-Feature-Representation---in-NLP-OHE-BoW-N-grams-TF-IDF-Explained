###One Hot Encoding (OHE) :

```python
movie is good      --> [movie, is, good]
movie is bad       --> [movie, is, bad]
movie is not good  --> [movie, is, not, good]

vocab = [movie, is, good, bad, not]

movie = [1, 0, 0, 0, 0]
good = [0, 0, 1, 0, 0]
bad = [0, 0, 0, 1, 0]
```

```python
1) Tokenization
2) create Vocab
3) create vectores : value of word is 1 else 0
```

```python
## Pros :
>> Simple & Intuative

## Cons :
>> Semantic meaning is not captured
>> vector size will increse with increase in vocab size
>> Sparsity :
>> OOV : Out of Vocab
```

###Bag-of-words / CountVectorizer :

```python
movie is good      --> [movie, is, good]
movie is bad       --> [movie, is, bad]
movie is not good  --> [movie, is, not, good]

vocab = [movie, is, good, bad, not]

>> movie is good good = [1, 1, 2, 0, 0]
>> movie is bad = [1, 1, 0, 1, 0]
>> movie is not good = [1, 1, 1, 0, 1]
```

```python
review = [
    "a movie is good good",
    "movie is bad",
    "movie is not good"
    ]
sentiment = [1, 0, 0]
```

```python
import pandas as pd

df = pd.DataFrame({"review":review, "sentiment":sentiment})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
vectors = cv.fit_transform(df["review"])
```

```python
vectors[0].toarray()
# movie is good good
```

```python
cv.vocabulary_
## {word : index}
```

```python
# movie is bad
vectors[1].toarray()
# [1, 0, 1, 1, 0]
```

```python
# cv = CountVectorizer()
## Hyperparameters :

lowercase : default=True
convert all token in loercase

stop_words : default=None --> "english", list

max_df : default=1.0 --> float, int
float(0.7)  --> Ignore token occuring in more than 70% documnets
int(10)  --> Ignore token occuring in more than 10 documents

min_df : default=1 --> float, int
float(0.1)  --> Ignore token occuring in less than 10% documnets
int(2)  --> Ignore token occuring in less than 2 documents

ngram_range : default=(1,1) --> (min, max)
```

```python
## Pros :
>> Simple :
>> Explainability :

## Cons :
>> Semantic meaning is not captured
>> vector size will increse with increase in vocab size
>> Sparsity :
>> OOV : Out of Vocab
```

###TF-IDF :

```python
TF : Term Frequency
IDF : Inverse Document Frequency
```

```python
TF = no. of times word occured in doc / total no. of words in doc

IDF = log(total no. of docs / no. of docs containing word)

TF-IDF = TF * IDF
```

```python
movie is good      --> [movie, is, good]
movie is bad       --> [movie, is, bad]
movie is not good  --> [movie, is, not, good]

vocab = [movie, is, good, bad, not]
```

```python
# movie is good
TF(movie) = 1 / 3
IDF(movie) = log(3 / 3)

TF-IDF(movie) = TF * IDF
```
